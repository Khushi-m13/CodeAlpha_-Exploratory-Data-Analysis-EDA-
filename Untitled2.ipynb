```python
# STEP 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# STEP 2: Read the uploaded CSV file
df = pd.read_csv('Netflix_Movies_and_TV_Shows.csv')
df.head()  # Show first 5 rows of data
```

```python
# Check info and missing values
df.info()
df.isnull().sum()
```

```python
# Check info and missing values
df.info()
df.isnull().sum()

# STEP 3: Drop rows with missing 'Genre' or 'Release Year'
# Based on typical Netflix datasets, the columns might be 'listed_in' and 'release_year'
# If these are still incorrect, check df.columns output for the correct names.
try:
    df = df.dropna(subset=['listed_in', 'release_year'])
except KeyError as e:
    print(f"Error dropping rows: {e}. Please check the actual column names in your DataFrame.")
    print("DataFrame columns:", df.columns.tolist())


# Display basic stats
df.describe(include='all')

# %% [markdown]
#

# %%
# STEP 4: Movies vs TV Shows

# Print the column names to identify the correct column (already done above now)
# print(df.columns)

# Replace 'Type' with the correct column name from the output above
plt.figure(figsize=(6,4))
# Replace 'Type' with the actual column name for show types (e.g., 'type', 'Show Type')
# Assuming 'type' is the correct column name for movie/tv show type
df['type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Movies vs TV Shows')
plt.xlabel('Type') # Update label if needed (maybe use 'type' here too for consistency)
plt.ylabel('Count')
plt.show()
```

```python
# STEP 3: Movies vs TV Shows

# Print the column names to identify the correct column
print(df.columns)

# Replace 'Type' with the correct column name from the output above
plt.figure(figsize=(6,4))
# Replace 'Type' with the actual column name for show types (e.g., 'type', 'Show Type')
df['type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Movies vs TV Shows')
plt.xlabel('Type') # Update label if needed (maybe use 'type' here too for consistency)
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['listed_in'].value_counts().head(10).plot(kind='barh', color='salmon')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()
```

```python
plt.figure(figsize=(10,5))
# Use the correct column name 'release_year'
df['release_year'].value_counts().sort_index().plot(kind='line', marker='o', color='green')
plt.title('Number of Titles Released Per Year')
# Update the xlabel to match the correct column name
plt.xlabel('release_year')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
# Change 'Country' to 'country'
df['country'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```
